```python
from google.colab import drive
import pandas as pd

# Montar Google Drive
drive.mount('/content/drive')

# Ruta a la carpeta de los archivos en tu Google Drive
file_paths = {
    "city_day": "/content/drive/MyDrive/AirQualityData/city_day.csv",
    "city_hour": "/content/drive/MyDrive/AirQualityData/city_hour.csv",
    "station_day": "/content/drive/MyDrive/AirQualityData/station_day.csv",
    "station_hour": "/content/drive/MyDrive/AirQualityData/station_hour.csv",
    "stations": "/content/drive/MyDrive/AirQualityData/stations.csv",
}

# Cargar los archivos desde Google Drive
dataframes = {name: pd.read_csv(path) for name, path in file_paths.items()}

# Regla para rellenar valores nulos en columnas numéricas
def fill_numeric_with_median(df, columns):
    for col in columns:
        if col in df.columns:
            df[col] = df[col].fillna(df[col].median())
    return df

# Aplicar reglas a los archivos
for name, df in dataframes.items():
    if name in ["city_day", "city_hour", "station_day", "station_hour"]:
        # Identificar columnas numéricas
        numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns

        # Rellenar columnas numéricas con la mediana
        df = fill_numeric_with_median(df, numeric_columns)

        # Rellenar `AQI_Bucket` con "Desconocido"
        if 'AQI_Bucket' in df.columns:
            df['AQI_Bucket'] = df['AQI_Bucket'].fillna("Desconocido")

        # Eliminar columna Xylene si existe
        if 'Xylene' in df.columns:
            df.drop(columns=['Xylene'], inplace=True)

    elif name == "stations":
        # Rellenar `Status` con "Desconocido"
        if 'Status' in df.columns:
            df['Status'] = df['Status'].fillna("Desconocido")

    # Guardar los cambios en Google Drive
    output_path = f"/content/drive/My Drive/AirQualityData/processed_{name}.csv"
    df.to_csv(output_path, index=False)
    print(f"Archivo procesado y guardado: {output_path}")


    # Rutas de los CSV originales procesados en Google Drive
file_paths = {
    "processed_city_day": "/content/drive/MyDrive/AirQualityData/processed_city_day.csv",
    "processed_city_hour": "/content/drive/MyDrive/AirQualityData/processed_city_hour.csv",
    "processed_station_day": "/content/drive/MyDrive/AirQualityData/processed_station_day.csv",
    "processed_station_hour": "/content/drive/MyDrive/AirQualityData/processed_station_hour.csv",
    "processed_stations": "/content/drive/MyDrive/AirQualityData/processed_stations.csv",
}

# Ruta donde se guardarán los archivos Parquet
parquet_paths = {
    "processed_city_day": "/content/drive/MyDrive/AirQualityData/refined_city_day.parquet",
    "processed_city_hour": "/content/drive/MyDrive/AirQualityData/refined_city_hour.parquet",
    "processed_station_day": "/content/drive/MyDrive/AirQualityData/refined_station_day.parquet",
    "processed_station_hour": "/content/drive/MyDrive/AirQualityData/refined_station_hour.parquet",
    "processed_stations": "/content/drive/MyDrive/AirQualityData/refined_stations.parquet",
}

# Convertir cada archivo CSV procesado a Parquet
for name, csv_path in file_paths.items():
    parquet_path = parquet_paths[name]
    df = pd.read_csv(csv_path)  # Cargar el CSV
    df.to_parquet(parquet_path, index=False)  # Guardar como Parquet
    print(f"Archivo convertido a Parquet: {parquet_path}")

```
